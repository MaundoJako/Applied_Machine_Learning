```python
import cv2
from cv2 import CascadeClassifier
from matplotlib.patches import Rectangle
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = plt.imread('/Users/jakemaund/Desktop/Applied_Machine_Learning/Data/family_old.jpeg')
fig, ax = plt.subplots(figsize=(12, 8), subplot_kw={'xticks': [], 'yticks':[]})
ax.imshow(image)

model = CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

faces = model.detectMultiScale(image, minNeighbors=10) # default=3, increasing means stricter face detection.

for face in faces:
    x, y, w, h = face
    rect = Rectangle((x, y), w, h, color='red', fill=False, lw=2)
    ax.add_patch(rect)
```
